```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# A = [S, I, R]
# dAdt = [dSdt, dIdt, dRdt]
def dAdt(A, t, beta, N ,gamma):
    S, I ,R = A
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return [dSdt, dIdt, dRdt]
```

```python
gamma = 0.1
N = 1.1*10**7
beta = 0.39

I0 = 574
S0 = N - I0
R0 = 0
```

```python
times = np.arange(0, 100, 1)
```

```python
sol = odeint(dAdt, [S0, I0, R0], times, args=(beta, N, gamma))
S = sol.T[0]
I = sol.T[1]
R = sol.T[2]
```

```python
plt.plot(times, S)
plt.plot(times, I)
plt.plot(times, R)
```

```python
import turtle
import random
```

```python
class Hunter:
    def __init__(self, screen):
        self.screen = screen
        self.turtle = turtle.Turtle()
        self.turtle.penup()
    
    def move(self, direction):
        dir_dict = {"up": 90, "down": 270, "left": 180, "right": 0}
        self.turtle.setheading(dir_dict[direction])
```

```python
class Dot:
    def __init__(self, screen):
        self.screen = screen
        self.x = random.randint(-200, 200)
        self.y = random.randint(-200, 200)
        self.turtle = turtle.Turtle()
        self.turtle.shape("circle")
        self.turtle.color("blue")
        self.turtle.penup()
        self.turtle.goto(self.x, self.y)
```

```python
screen = turtle.Screen()
screen.setup(width=800, height=600)
screen.title("Seek the Dot")
screen.bgcolor("white")
screen.tracer(0)

hunter = Hunter(screen)
dot = Dot(screen)

def tick():
    # ... hier Spielzustand aktualisieren ...
    screen.update()
    screen.ontimer(tick, 16)  # ~60 FPS
 
screen.listen()
tick()
screen.mainloop()
```

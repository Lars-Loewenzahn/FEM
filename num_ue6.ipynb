### 1.1.

$N_k(x)=\prod_{j=0}^{k-1}\left(x-x_j\right) \text { für } k=1, \ldots, n \text {. }$

$
\begin{aligned}
&\text { Das } k \text {-te Newton-Polynom } N_k(x) \text { hat Grad } k \text {, da es als Produkt von } k \text { linearen Faktoren definiert ist: }\\
&N_k(x)=\left(x-x_0\right)\left(x-x_1\right) \cdots\left(x-x_{k-1}\right) .
\end{aligned}$

- $\quad N_0(x)$ hat Grad 0,
- $\quad N_1(x)$ hat Grad 1 ,
- $\quad N_2(x)$ hat Grad 2,
- ...
- $\quad N_n(x)$ hat Grad $n$.

Da alle N von einem anderem Grad sind folgt unabhängigkeit. 
a_0 N_0(x)+a_1 N_1(x)+\cdots+a_n N_n(x)=0 .

Daraus folgt auch, dass die Basis $\left\{N_0, \ldots, N_n\right\}$ eine Basis von $\Pi_n$ ist.
$$P(x)=a_0 N_0(x)+a_1 N_1(x)+\cdots+a_n N_n(x)$$

```python
import matplotlib.pyplot as plt
import random

# X-Werte von 0 bis 10
x_values = list(range(11))

# Y-Werte zufällig zwischen 11 und 19
y_values = [random.randint(11, 19) for _ in x_values]
```

```python
# Plot erstellen
plt.figure(figsize=(8, 5))
plt.scatter(x_values, y_values, color='lightblue', label='Datenpunkte')
plt.title('Varianz vs. Standardabweichung')
plt.xlabel('X-Achse (0 bis 10)')
# plt.ylabel('Y-Achse (-1 bis 31)')
plt.ylim(-1, 31)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)

# Grafik anzeigen
plt.show()
```

```python
mue = 0
for i in range(len(x_values)):
    mue += y_values[i] / len(x_values)
print(mue)

# Plot erstellen
plt.figure(figsize=(8, 5))
plt.scatter(x_values, y_values, color='lightblue', label='Datenpunkte')
plt.plot(x_values, [mue for _ in x_values], color='red', label='$E(x)$')
plt.title('Varianz vs. Standardabweichung')
plt.xlabel('X-Achse (0 bis 10)')
# plt.ylabel('Y-Achse (-1 bis 31)')
plt.ylim(-1, 31)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)

# Grafik anzeigen
plt.show()
```

```python
y_cleaned = [mue + abs(y_value - mue) for y_value in y_values]
# Plot erstellen
plt.figure(figsize=(8, 5))
plt.scatter(x_values, y_values, color='lightblue', label='Datenpunkte')
plt.scatter(x_values, y_cleaned, color='blue', label='Abweichung + $E(x)$')
plt.plot(x_values, [mue for _ in x_values], color='red', label='$E(x)$')
plt.title('Varianz vs. Standardabweichung')
plt.xlabel('X-Achse (0 bis 10)')
# plt.ylabel('Y-Achse (-1 bis 31)')
plt.ylim(-1, 31)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)

# Grafik anzeigen
plt.show()
```

```python
y_vars = [mue + abs(y_value - mue)**2 for y_value in y_values]

# Plot erstellen
plt.figure(figsize=(8, 5))
plt.scatter(x_values, y_values, color='lightblue', label='Datenpunkte')
plt.scatter(x_values, y_cleaned, color='blue', label='Abweichung + $E(x)$')
plt.scatter(x_values, y_vars, color='yellow', label='Abweichung² + $E(x)$')
plt.plot(x_values, [mue for _ in x_values], color='red', label='$E(x)$')
plt.title('Varianz vs. Standardabweichung')
plt.xlabel('X-Achse (0 bis 10)')
# plt.ylabel('Y-Achse (-1 bis 31)')
plt.ylim(-1, 31)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)

# Grafik anzeigen
plt.show()
```

```python
# Varianz berechnen
varianz = 0
for i in range(len(x_values)):
    varianz += (y_values[i] - mue)**2 / len(x_values)

# Plot erstellen
plt.figure(figsize=(8, 5))
plt.scatter(x_values, y_values, color='lightblue', label='Datenpunkte')
plt.scatter(x_values, y_cleaned, color='blue', label='Abweichung + $E(x)$')
plt.scatter(x_values, y_vars, color='yellow', label='Abweichung² + $E(x)$')
plt.plot(x_values, [mue + varianz for _ in x_values], color='yellow', label=r'$ E(x) + Var(x)$')
plt.plot(x_values, [mue for _ in x_values], color='red', label='$E(x)$')
plt.title('Varianz vs. Standardabweichung')
plt.xlabel('X-Achse (0 bis 10)')
# plt.ylabel('Y-Achse (-1 bis 31)')
plt.ylim(-1, 31)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)

# Grafik anzeigen
plt.show()
```

```python
sigma = np.sqrt(varianz)

# Plot erstellen
plt.figure(figsize=(8, 5))
plt.scatter(x_values, y_values, color='lightblue', label='Datenpunkte')
plt.scatter(x_values, y_cleaned, color='blue', label='Abweichung + $E(x)$')
plt.scatter(x_values, y_vars, color='yellow', label='Abweichung² + $E(x)$')
plt.plot(x_values, [mue + varianz for _ in x_values], color='yellow', label=r'$ E(x) + Var(x)$')
plt.plot(x_values, [mue + sigma for _ in x_values], color='blue', label=r'$ E(x) + \sigma $')
plt.plot(x_values, [mue for _ in x_values], color='red', label='$E(x)$')
plt.title('Varianz vs. Standardabweichung')
plt.xlabel('X-Achse (0 bis 10)')
# plt.ylabel('Y-Achse (-1 bis 31)')
plt.ylim(-1, 31)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)

# Grafik anzeigen
plt.show()
```

```python
# 4. Streudiagramm: Größe vs. Gewicht von Schülern
height = np.random.normal(170, 10, 50)  # Random heights (mean=170, std=10)
weight = np.random.normal(70, 15, 50)  # Random weights (mean=70, std=15)

plt.figure(figsize=(8, 6))
plt.scatter(height, weight, color='purple')
plt.title('Größe vs. Gewicht von Schülern')
plt.xlabel('Größe (cm)')
plt.ylabel('Gewicht (kg)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.stats import linregress

# Daten für Größe und Gewicht von Schülern
# np.random.seed(0)  # Seed for reproducibility
# height = np.random.normal(170, 10, 50)  # Random heights (mean=170, std=10)
# weight = np.random.normal(70, 15, 50)  # Random weights (mean=70, std=15)

# Berechnung der Erwartungswerte und Standardabweichungen
mean_height = np.mean(height)
std_height = np.std(height)
mean_weight = np.mean(weight)
std_weight = np.std(weight)

# Regressionslinie
slope, intercept, _, _, _ = linregress(height, weight)
regression_line = slope * height + intercept

# Erstellung des Streudiagramms mit zusätzlichen Informationen
plt.figure(figsize=(8, 6))
plt.scatter(height, weight, color='purple', label='Datenpunkte')
plt.axvline(mean_height, color='green', linestyle='--', label=f'Erwartungswert Größe: {mean_height:.1f} cm')
plt.axhline(mean_weight, color='blue', linestyle='--', label=f'Erwartungswert Gewicht: {mean_weight:.1f} kg')
plt.fill_betweenx([mean_weight - std_weight, mean_weight + std_weight], height.min(), height.max(),
                  color='blue', alpha=0.1, label='Standardabweichung Gewicht')
plt.fill_between([mean_height - std_height, mean_height + std_height], weight.min(), weight.max(),
                  color='green', alpha=0.1, label='Standardabweichung Größe')
plt.plot(height, regression_line, color='red', label='Regressionslinie')

# Titel, Achsenbeschriftungen und Legende
plt.title('Größe vs. Gewicht von Schülern (mit Zusatzinformationen)')
plt.xlabel('Größe (cm)')
plt.ylabel('Gewicht (kg)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Wertebereich für x
x = np.linspace(0, 10, 500)

# 1. Normalform
a1, k = 2, 0.5
f1 = a1 * np.exp(k * x)

# 2. Halbwertsform
a2, T_half = 100, 5
f2 = a2 * (0.5) ** (x / T_half)

# 3. Verdoppelungsform
a3, T_double = 1, 3
f3 = a3 * (2) ** (x / T_double)

# 4. Zinseszinsform
K0, r = 1000, 0.05
f4 = K0 * (1 + r) ** x

# Plotten der Funktionen
plt.figure(figsize=(12, 8))

# Normalform
plt.plot(x, f1, label=r'$f(x) = 2 \cdot e^{0.5 \cdot x}$', color='blue')

# Halbwertsform
plt.plot(x, f2, label=r'$f(x) = 100 \cdot \left(\frac{1}{2}\right)^{x/5}$', color='green')

# Verdoppelungsform
plt.plot(x, f3, label=r'$f(x) = 1 \cdot 2^{x/3}$', color='orange')

# Zinseszinsform
plt.plot(x, f4, label=r'$f(x) = 1000 \cdot (1 + 0.05)^x$', color='red')

# Plot-Details
plt.title("Exponentialfunktionen in verschiedenen Formen", fontsize=16)
plt.xlabel("x", fontsize=14)
plt.ylabel("f(x)", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Werte für x und die Wachstumsrate
x = np.arange(0, 10, 1)  # Schritte
a = 1.5  # Wachstumsrate
b = 2  # Anfangswert

# Berechnung der Werte y = b * a^x
y = b * a**x

# Plot erstellen
plt.figure(figsize=(10, 6))

# Wachstumskurve
plt.plot(x, y, marker='o', label=fr"$y = {b} \cdot {a}^x$")

# Markierungen für den Faktor
for i in range(len(x) - 1):
    plt.annotate(fr"$\times {a}$", 
                 xy=(x[i] + 0.5, (y[i] + y[i + 1]) / 2), 
                 textcoords="offset points", 
                 xytext=(0, 10), ha='center', fontsize=10, color='blue')

# Titel und Achsenbeschriftungen
plt.title("Verdeutlichung des Wachstumsfaktors", fontsize=16)
plt.xlabel("Schritte (x)", fontsize=14)
plt.ylabel("Wert (y)", fontsize=14)

# Legende und Gitter
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Grafik anzeigen
plt.show()
```

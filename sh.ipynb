```python
import numpy as np
import matplotlib.pyplot as plt

# Datenpunkte (x-Werte) und Wahrscheinlichkeiten (p-Werte)
x = np.array([1, 2, 3, 4, 5])
p = np.array([0.1, 0.2, 0.4, 0.2, 0.1])

# Plot erstellen
fig, ax = plt.subplots(figsize=(8, 5))

# Datenpunkte plotten
ax.scatter(x, np.zeros_like(x), color='blue', label='Datenpunkte (x)')

# Achseneinstellungen
ax.set_title("Datenpunkte")
ax.set_xlabel("Werte (x)")
ax.set_ylabel("Abweichungen")
ax.axhline(0, color='black', linewidth=0.5)
ax.legend()
ax.grid(True, alpha=0.5)

plt.show()
```

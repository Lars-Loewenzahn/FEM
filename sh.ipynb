```python
import numpy as np
import matplotlib.pyplot as plt

# Datenpunkte (x-Werte) und Wahrscheinlichkeiten (p-Werte)
x = np.array([1, 2, 3, 4, 5])
p = np.array([0.1, 0.2, 0.4, 0.2, 0.1])

# Plot erstellen
fig, ax = plt.subplots(figsize=(8, 5))

# Datenpunkte plotten
ax.scatter(x, np.zeros_like(x), color='blue', label='Datenpunkte (x)')

# Achseneinstellungen
ax.set_title("Datenpunkte")
ax.set_xlabel("Werte (x)")
ax.set_ylabel("Abweichungen")
ax.axhline(0, color='black', linewidth=0.5)
ax.legend()
ax.grid(True, alpha=0.5)

plt.show()
```

```python
import matplotlib.pyplot as plt
import random

# X-Werte von 0 bis 10
x_values = list(range(11))

# Y-Werte zuf√§llig zwischen 11 und 19
y_values = [random.randint(11, 19) for _ in x_values]
```

```python
# Plot erstellen
plt.figure(figsize=(8, 5))
plt.scatter(x_values, y_values, color='lightblue', label='Datenpunkte')
plt.title('Varianz vs. Standardabweichung')
plt.xlabel('X-Achse (0 bis 10)')
# plt.ylabel('Y-Achse (-1 bis 31)')
plt.ylim(-1, 31)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)

# Grafik anzeigen
plt.show()
```

```python
mue = 0
for i in range(len(x_values)):
    mue += y_values[i] / len(x_values)
print(mue)

# Plot erstellen
plt.figure(figsize=(8, 5))
plt.scatter(x_values, y_values, color='lightblue', label='Datenpunkte')
plt.plot(x_values, [mue for _ in x_values], color='red', label='$E(x)$')
plt.title('Varianz vs. Standardabweichung')
plt.xlabel('X-Achse (0 bis 10)')
# plt.ylabel('Y-Achse (-1 bis 31)')
plt.ylim(-1, 31)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)

# Grafik anzeigen
plt.show()
```

```python
y_cleaned = [mue + abs(y_value - mue) for y_value in y_values]
# Plot erstellen
plt.figure(figsize=(8, 5))
plt.scatter(x_values, y_values, color='lightblue', label='Datenpunkte')
plt.scatter(x_values, y_cleaned, color='blue', label='Abweichung + $E(x)$')
plt.plot(x_values, [mue for _ in x_values], color='red', label='$E(x)$')
plt.title('Varianz vs. Standardabweichung')
plt.xlabel('X-Achse (0 bis 10)')
# plt.ylabel('Y-Achse (-1 bis 31)')
plt.ylim(-1, 31)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)

# Grafik anzeigen
plt.show()
```

```python
y_vars = [mue + abs(y_value - mue)**2 for y_value in y_values]

# Plot erstellen
plt.figure(figsize=(8, 5))
plt.scatter(x_values, y_values, color='lightblue', label='Datenpunkte')
plt.scatter(x_values, y_cleaned, color='blue', label='Abweichung + $E(x)$')
plt.scatter(x_values, y_vars, color='yellow', label='Abweichung¬≤ + $E(x)$')
plt.plot(x_values, [mue for _ in x_values], color='red', label='$E(x)$')
plt.title('Varianz vs. Standardabweichung')
plt.xlabel('X-Achse (0 bis 10)')
# plt.ylabel('Y-Achse (-1 bis 31)')
plt.ylim(-1, 31)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)

# Grafik anzeigen
plt.show()
```

```python
# Varianz berechnen
varianz = 0
for i in range(len(x_values)):
    varianz += (y_values[i] - mue)**2 / len(x_values)

# Plot erstellen
plt.figure(figsize=(8, 5))
plt.scatter(x_values, y_values, color='lightblue', label='Datenpunkte')
plt.scatter(x_values, y_cleaned, color='blue', label='Abweichung + $E(x)$')
plt.scatter(x_values, y_vars, color='yellow', label='Abweichung¬≤ + $E(x)$')
plt.plot(x_values, [mue + varianz for _ in x_values], color='yellow', label=r'$ E(x) + Var(x)$')
plt.plot(x_values, [mue for _ in x_values], color='red', label='$E(x)$')
plt.title('Varianz vs. Standardabweichung')
plt.xlabel('X-Achse (0 bis 10)')
# plt.ylabel('Y-Achse (-1 bis 31)')
plt.ylim(-1, 31)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)

# Grafik anzeigen
plt.show()
```

```python
sigma = np.sqrt(varianz)

# Plot erstellen
plt.figure(figsize=(8, 5))
plt.scatter(x_values, y_values, color='lightblue', label='Datenpunkte')
plt.scatter(x_values, y_cleaned, color='blue', label='Abweichung + $E(x)$')
plt.scatter(x_values, y_vars, color='yellow', label='Abweichung¬≤ + $E(x)$')
plt.plot(x_values, [mue + varianz for _ in x_values], color='yellow', label=r'$ E(x) + Var(x)$')
plt.plot(x_values, [mue + sigma for _ in x_values], color='blue', label=r'$ E(x) + \sigma $')
plt.plot(x_values, [mue for _ in x_values], color='red', label='$E(x)$')
plt.title('Varianz vs. Standardabweichung')
plt.xlabel('X-Achse (0 bis 10)')
# plt.ylabel('Y-Achse (-1 bis 31)')
plt.ylim(-1, 31)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)

# Grafik anzeigen
plt.show()
```

```python
# 2. Balkendiagramm: Fehlstunden nach Jahrgangsstufe
grades = ['9. Klasse', '10. Klasse', '11. Klasse', '12. Klasse', '13. Klasse']
absences = [5, 8, 10, 15, 12]

plt.figure(figsize=(8, 6))
plt.bar(grades, absences, color='skyblue')
plt.title('Fehlstunden nach Jahrgangsstufe')
plt.xlabel('Jahrgangsstufe')
plt.ylabel('Fehlstunden')
plt.show()
```

```python
# Daten f√ºr das Liniendiagramm mit zwei Datens√§tzen
semesters = ['1. Semester', '2. Semester', '3. Semester', '4. Semester']
average_grades_set1 = [3.2, 3.0, 2.8, 2.5]
average_grades_set2 = [3.5, 3.3, 3.0, 2.7]

# Erstellung des Liniendiagramms
plt.figure(figsize=(8, 6))
plt.plot(semesters, average_grades_set1, marker='o', color='green', label='Gruppe A')
plt.plot(semesters, average_grades_set2, marker='s', color='blue', label='Gruppe B')

# Farblich markierter Bereich zwischen den Linien
plt.fill_between(
    semesters,
    average_grades_set1,
    average_grades_set2,
    color='lightgray',
    alpha=0.5,
    label='Differenzbereich'
)

# Hinzuf√ºgen von Beschriftungen und Legende
plt.title('Durchschnittsnoten √ºber Semester')
plt.xlabel('Semester')
plt.ylabel('Durchschnittsnote')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 4. Streudiagramm: Gr√∂√üe vs. Gewicht von Sch√ºlern
height = np.random.normal(170, 10, 50)  # Random heights (mean=170, std=10)
weight = np.random.normal(70, 15, 50)  # Random weights (mean=70, std=15)

plt.figure(figsize=(8, 6))
plt.scatter(height, weight, color='purple')
plt.title('Gr√∂√üe vs. Gewicht von Sch√ºlern')
plt.xlabel('Gr√∂√üe (cm)')
plt.ylabel('Gewicht (kg)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.stats import linregress

# Daten f√ºr Gr√∂√üe und Gewicht von Sch√ºlern
# np.random.seed(0)  # Seed for reproducibility
# height = np.random.normal(170, 10, 50)  # Random heights (mean=170, std=10)
# weight = np.random.normal(70, 15, 50)  # Random weights (mean=70, std=15)

# Berechnung der Erwartungswerte und Standardabweichungen
mean_height = np.mean(height)
std_height = np.std(height)
mean_weight = np.mean(weight)
std_weight = np.std(weight)

# Regressionslinie
slope, intercept, _, _, _ = linregress(height, weight)
regression_line = slope * height + intercept

# Erstellung des Streudiagramms mit zus√§tzlichen Informationen
plt.figure(figsize=(8, 6))
plt.scatter(height, weight, color='purple', label='Datenpunkte')
plt.axvline(mean_height, color='green', linestyle='--', label=f'Erwartungswert Gr√∂√üe: {mean_height:.1f} cm')
plt.axhline(mean_weight, color='blue', linestyle='--', label=f'Erwartungswert Gewicht: {mean_weight:.1f} kg')
plt.fill_betweenx([mean_weight - std_weight, mean_weight + std_weight], height.min(), height.max(),
                  color='blue', alpha=0.1, label='Standardabweichung Gewicht')
plt.fill_between([mean_height - std_height, mean_height + std_height], weight.min(), weight.max(),
                  color='green', alpha=0.1, label='Standardabweichung Gr√∂√üe')
plt.plot(height, regression_line, color='red', label='Regressionslinie')

# Titel, Achsenbeschriftungen und Legende
plt.title('Gr√∂√üe vs. Gewicht von Sch√ºlern (mit Zusatzinformationen)')
plt.xlabel('Gr√∂√üe (cm)')
plt.ylabel('Gewicht (kg)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 5. Boxplot: Verteilung der Testnoten in einer Klasse
test_scores = [np.random.normal(75, 10, 30) for _ in range(5)]

plt.figure(figsize=(8, 6))
plt.boxplot(test_scores, labels=['Test 1', 'Test 2', 'Test 3', 'Test 4', 'Test 5'])
plt.title('Verteilung der Testnoten in einer Klasse')
plt.xlabel('Tests')
plt.ylabel('Punktzahl')
plt.show()
```

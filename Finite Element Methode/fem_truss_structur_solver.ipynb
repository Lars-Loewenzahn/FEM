```python
import matplotlib.pyplot as plt
```

```python
class Truss:
    def __init__(self, node1, node2, E, A, I):
        self.node1 = node1
        self.node2 = node2
        self.youngs_modulus = E
        self.crosssectional_aera = A
        self.area_moment = I

    def get_length(self):
        return self.node1.get_distance(self.node2)
    
    def get_line(self):
        x = [self.node1.x, self.node2.x]
        y = [self.node1.y, self.node2.y]
        return x ,y
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Linie zeichnen
plt.plot(x, y)

# Titel und Achsenbeschriftungen
plt.title("Einfache Linie")
plt.xlabel("x-Achse")
plt.ylabel("y-Achse")

# Anzeige
plt.show()
```

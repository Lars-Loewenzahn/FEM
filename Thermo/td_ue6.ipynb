## Aufgabe 6.1


In einer Warmwasserheizung soll isobar ein Wassermassenstrom $\dot{m}_{\mathrm{W}}$ von $T_{1 \mathrm{~W}}=30^{\circ} \mathrm{C}$ auf $T_{2 \mathrm{~W}}=50^{\circ} \mathrm{C}$ isobar aufgeheizt werden. Für die Erwärmung werden die Abgase eines Ölbrenners verwendet, welche als ideales Gas betrachtet werden können.


a) Wie groß muss $\dot{m}_{\mathrm{W}}$ für die Übertragung eines Wärmestroms von 10 kW sein?


b) Der Abgasstrom kühlt von $T_{1 \mathrm{~A}}=800^{\circ} \mathrm{C}$ auf $T_{2 \mathrm{~A}}=90^{\circ} \mathrm{C}$ ab. Wie groß ist der Abgasmassenstrom?



$$
\begin{aligned}
& \text { Daten des Wassers: }\left.c_p\right|_{0^{\circ} \mathrm{C}} ^{50 \mathrm{C}}=4.183 \mathrm{~kJ} /(\mathrm{kg} \mathrm{~K}) \\
& \left.\underline{\text { Daten des Abgases: }} c_p^{\circ}\right|_{0^{\circ} \mathrm{C}} ^{9 \mathrm{C}^{\circ}}=1.007 \mathrm{~kJ} /(\mathrm{kg} \mathrm{~K}),\left.c_p^{\circ}\right|_{0^{\circ} \mathrm{C}} ^{800^{\circ} \mathrm{C}}=1.071 \mathrm{~kJ} /(\mathrm{kg} \mathrm{~K})
\end{aligned}
$$

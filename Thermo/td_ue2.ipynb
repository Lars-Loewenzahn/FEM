## Aufgabe 2.1


Wir betrachten die Edelgase Helium (Molmasse $4 \mathrm{~g} / \mathrm{mol}$ ) und Argon (Molmasse $40 \mathrm{~g} / \mathrm{mol}$ ). Für beide Gase nutzen wir das Idealgas-Modell.

a) Berechnen Sie die molare innere Energie von Helium und Argon bei den beiden Temperaturen $T_1=300 \mathrm{~K}$ und $T_2=400 \mathrm{~K}$.

```python
# Beides sind einatomige Gase
# U ist die molare innere Energie 
# für einatomige Gase gilt U = 3/2 RT
# für zweiatomige Gase gilt U = 5/2 RT
# für mehratomige Gase gilt U = f/2 RT - f = Freiheitsgrade

R = 8.314472  # J/molK universelle Gaskonstante
T_1 = 300  # K
T_2 = 400  # K
U_1 = 3 / 2 * R * T_1
U_2 = 3 / 2 * R * T_2

print("U_1:", U_1, "J/mol")
print("U_2:", U_2, "J/mol")
```

b) Wie groß ist die spezifische innere Energie $u$ in den genannten Fällen?

```python
# die speziische innere Energie ist u
# u = U/M 
# M ist die Molare Masse
M_helium = 4 #g/mol
M_argon = 40 # g/mol

u_helium_1 = U_1 / M_helium
print("u_helium_1: ", u_helium_1, "kJ/kg")
u_helium_2 = U_2 / M_helium
print("u_helium_2: ", u_helium_2, "kJ/kg")
u_argon_1 = U_1 / M_argon
print("u_argon_1: ", u_argon_1, "kJ/kg")
u_argon_2 = U_2 / M_argon
print("u_argon_2: ", u_argon_2, "kJ/kg")
```

c) Wie groß ist jeweils die mittlere Geschwindigkeit der Teilchen?

Hinweis: $R_{\mathrm{m}}=8.314472 \mathrm{~J} /(\mathrm{mol} \mathrm{K}), N_{\mathrm{A}}=6.0225 \cdot 10^{23}$ Teilchen $/ \mathrm{mol}$

```python
N = 6.02225 * 10**23
E_kin_1 = U_1 / N
E_kin_2 = U_2 / N
# print("E_kin_1: ", E_kin_1, "kJ/kg")
# print("E_kin_2: ", E_kin_2, "kJ/kg")
m_helium = M_helium*0.001/N
m_argon = M_argon*0.001/N

v_helium_1 = (2*E_kin_1/m_helium)**0.5
print("v_helium_1: ", v_helium_1, "m/s")
v_helium_2 = (2*E_kin_2/m_helium)**0.5
print("v_helium_2: ", v_helium_2, "m/s")
v_argon_1 = (2*E_kin_1/m_argon)**0.5
print("v_argon_1: ", v_argon_1, "m/s")
v_argon_2 = (2*E_kin_2/m_argon)**0.5
print("v_argon_2: ", v_argon_2, "m/s")
```

## Aufgabe 2.3

Bei einer Temperatur von $T_0=15^{\circ} \mathrm{C}$ wird am Erdboden ein Druck $p_0$ von 740 mm Quecksilbersäule ( 986.6 hPa ) gemessen. Berechnen Sie den Atmosphärendruck in einer Höhe von 2000 m unter den folgenden Annahmen:

a) die Temperatur sei konstant.

```python
import math
R = 0.287 #kJ / (kg K)
g = 9.81 # m / s²
p_0 = 98.66 # kPa
T_0 = 273.15 + 15
# Lösungen von ChatGPT: 
# p_1 = p_0*math.exp(-(g*M*h)/(R*T_0))
# p_1 = p_0*(1-(L*h)/T_0)**(g*M/(R*L))

# Musterlösung:
# Ein Kräftegleichgewicht wird aufgestellt. 
```

$$\Sigma F_i = 0 $$

$$ 0 = p*A - (p- dp)*A - \rho A g dz$$

$$ dp = - \rho g dz$$

$$ p \cdot v = RT$$

$$ \rho = \frac{p}{RT}$$

$$ dp = -p \cdot \frac{g}{RT} dz$$

Trennung der Variablen

$$  \frac{1}{p} dp = - \frac{g}{RT} dz $$

```python
import math
R = 287 #J / (kg K)
g = 9.81 # m / s²
p_0 = 98.66 # kPa
T_0 = 273.15 + 15
h = 2000
p_1  = p_0*math.exp(-(g*h)/(R*T_0))
print(p_1, " kPA")
```

b) der Gradient der Lufttemperatur sei konstant und betrage $6.5 \mathrm{~K} / \mathrm{km}$


Hinweis: Luft soll als ideales Gas betrachtet werden, mit $R=0.287 \mathrm{~kJ} /(\mathrm{kg} \mathrm{K})$. Die Erdbeschleunigung beträgt $g=9.81 \mathrm{~m} / \mathrm{s}^2$.

```python
b = 0.0065 # K / m
p_1 = p_0*((T_0-b*h)/T_0)**(g/(R*b))
print(p_1, " kPA")
```

Ein starrer verschlossener Behälter ist mit einem Fluid von $T=20^{\circ} \mathrm{C}$ und $p=1$ bar gefïllt. Berechnen Sie die Änderung des Drucks im Behälter bei einer Aufheizung des Fluides um 10 K , wenn es sich um folgende Fluide handelt:

a) Wasser

b) Quecksilber

c) Luft

Hinweis: Der (isobare) thermische Ausdehnungskoeffizient $\alpha$ und die isotherme Kompressibilität $\gamma$ von Wasser und Quecksilber sollen näherungsweise als druck- und temperaturunabhängig betrachtet werden. Luft soll als ideales Gas betrachtet werden.
\begin{tabular}{|cl|c|c|}
\hline & & Wasser & Quecksilber \\
\hline$\alpha$ & {$\left[\mathrm{K}^{-1}\right]$} & $18.0 \cdot 10^{-5}$ & $18.1 \cdot 10^{-5}$ \\
\hline$\gamma$ & {$\left[\mathrm{bar}^{-1}\right]$} & $45.4 \cdot 10^{-6}$ & $3.86 \cdot 10^{-6}$ \\
\hline
\end{tabular}

```python
alpha_water = 18*10**-5
alpha_merc = 18.1*10**-5
gamma_water = 45.4*10**-6
gamma_merc = 3.86*10**-6
T_0 = 273.15  + 20 # K
p_0 = 1 # bar 
dT = 10
# a 
alpha = alpha_water
gamma = gamma_water
p_1 = p_0  + alpha*dT/gamma
print(p_1, "bar")


# b
alpha = alpha_merc
gamma = gamma_merc
p_1 = p_0  + alpha*dT/gamma
print(p_1, "bar")

# c ist am einfachsten zu lösen, da es sich um ein Ideales Gas handelt.
T_1 = T_0 + dT

p_1 = T_1 * p_0/T_0 
print(p_1, "bar")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Gegebene Daten
p1 = 8  # bar
p2 = 2.83  # bar
p3 = 2  # bar

v1 = 210  # m/kg
v2 = 420  # m/kg
v3 = v2

kappa = 1.5

# Adiabatische Expansion pV^k = const
v_values = np.linspace(v1, v2, 100)
p_values = p1 * (v1 / v_values) ** kappa

# Plot erstellen
plt.figure(figsize=(8, 6))

# Adiabatische Expansion (Kurve)
plt.plot(v_values, p_values, label='Adiabat (1  2)', color='blue')

# Isochore Abkhlung (vertikale Linie)
plt.plot([v2, v3], [p2, p3], label='Isochor (2  3)', color='red', linestyle='--')

# Punkte fr die Zustnde
plt.plot(v1, p1, 'bo', label='Zustand 1 (8 bar, 210 m/kg)')
plt.plot(v2, p2, 'go', label='Zustand 2 (2.83 bar, 420 m/kg)')
plt.plot(v3, p3, 'ro', label='Zustand 3 (2 bar, 420 m/kg)')

# Schattierung der Volumennderungsarbeit
plt.fill_between(v_values, 0, p_values, color='lightgray', alpha=0.5, label='Volumennderungsarbeit')

# Achsenbeschriftung und Titel
plt.xlabel('Spezifisches Volumen $v$ (m/kg)')
plt.ylabel('Druck $p$ (bar)')
plt.title('$p-v$ Diagramm mit Volumennderungsarbeit')
plt.legend()
plt.grid(True)

# Diagramm anzeigen
plt.show()
```

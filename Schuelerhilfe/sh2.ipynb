```python
import os
import pandas as pd
import shutil
```

```python
column_srt = 'Name SRT'
column_txt = 'Name TXT'
input_folder_srt = 'C:\\Users\\Lars\\Documents\\GitHub\\FEM\\Schuelerhilfe\\input\\srts'
input_folder_txt = 'C:\\Users\\Lars\\Documents\\GitHub\\FEM\\Schuelerhilfe\\input\\txts'
output_folder_srt = 'C:\\Users\\Lars\\Documents\\GitHub\\FEM\\Schuelerhilfe\\output\\srts'
output_folder_txt = 'C:\\Users\\Lars\\Documents\\GitHub\\FEM\\Schuelerhilfe\\output\\txts'
first_file_srt = False
first_file_txt = False
start_index_srt = 402 # immer 2 weniger angeben
start_index_txt = 402 # immer 2 weniger angeben
read_up_srt = True
read_up_txt = True
```

```python
start_index = start_index_txt
column = column_txt
read_up = read_up_txt
import pandas as pd
df = pd.read_excel('C:\\Users\\Lars\\Documents\\GitHub\\FEM\\Schuelerhilfe\\bennenungen.xlsx')
new_names = []
num_rows = df.shape[0]  # Get the number of rows in the DataFrame

for i in range(10000):
    index = start_index + i if read_up else start_index - i
    if 0 <= index < num_rows:  # Check if the index is within bounds
        new_name = str(df.iloc[index][column])
        # print(new_name)
        new_names.append(new_name)
    else:
        break  # Exit the loop if the index is out of bounds

print(len(new_names))
```

```python
input_folder = input_folder_txt
output_folder = output_folder_txt
first_file = first_file_txt
start_index = start_index_txt
column = column_txt
read_up = read_up_txt

# Die Datei mit den Umbenennungen einlesen
df = pd.read_excel('C:\\Users\\Lars\\Documents\\GitHub\\FEM\\Schuelerhilfe\\bennenungen.xlsx')
```

```python
# Alle Dateien in dem Ordner holen
files = os.listdir(input_folder)
files = [file for file in files]

# Die Dateien sortieren
if first_file:
    files.sort(key=lambda x: os.path.getctime(os.path.join(input_folder, x)))
else:
    files.sort(key=lambda x: os.path.getctime(os.path.join(input_folder, x)), reverse=True)

print(len(files))
```

```python
# Clear the output directory
for filename in os.listdir(output_folder):
    file_path = os.path.join(output_folder, filename)
    try:
        if os.path.isfile(file_path) or os.path.islink(file_path):
            os.unlink(file_path)
        elif os.path.isdir(file_path):
            shutil.rmtree(file_path)
    except Exception as e:
        print(f'Failed to delete {file_path}. Reason: {e}')
```

```python
# Rename the files
for i in range(len(files)):
    old_name = files[i]
    new_name = new_names[i]
    print("Old name: " + old_name + "  New name:" + new_name)
    os.rename(os.path.join(input_folder, old_name), os.path.join(output_folder, new_name))
```

```python
if not new_extension.startswith('.'):
    new_extension = '.' + new_extension

# Iterate over all files in the directory
for filename in os.listdir(directory):
    # Construct the full file path
    file_path = os.path.join(directory, filename)

    # Check if it's a file (not a directory)
    if os.path.isfile(file_path):
        # Split the filename and its extension
        base_name, _ = os.path.splitext(filename)

        # Construct the new file name with the new extension
        new_file_name = base_name + new_extension

        # Construct the full new file path
        new_file_path = os.path.join(directory, new_file_name)

        # Rename the file
        os.rename(file_path, new_file_path)
        print(f'Renamed: {filename} to {new_file_name}')

# Example usage:
# change_file_extension('C:\\Users\\Lars\\Documents\\GitHub\\FEM\\Schuelerhilfe\\output\\srts', '.txt')
```

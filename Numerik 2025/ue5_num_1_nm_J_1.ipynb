```python
import math
def dy(t, y):
    return math.cos(t)*y
```

```python
def y(t):
    return np.exp(np.cos(t))
```

```python
import numpy as np
import matplotlib as mpl   
import matplotlib.pyplot as plt
```

```python
def loeservergleich(h):
    plt.figure()
    n = int(50//h)
    t = np.linspace(0, 50, n)
    plt.plot(t, y(t))

    y0 = 1
    y_euler = []
    y_euler.append(y0)
    for i in range(n-1):
        y_n = y_euler[-1] + h*dy(t[i], y_euler[-1])
        y_euler.append(y_n)
    plt.plot(t, y_euler)

    y_coll = []
    y_coll.append(y0)
    for i in range(n-1):
        y_n = y_coll[-1] + h*dy(t[i] + h/2, y_coll[-1] + h/2*dy(t[i], y_coll[-1]))
        y_coll.append(y_n)
    plt.plot(t, y_coll)

    y_heun = []
    y_heun.append(y0)
    for i in range(n-1):
        y_a = y_heun[-1] + h*dy(t[i], y_heun[-1])
        y_n = y_heun[-1] + h/2*(dy(t[i], y_heun[-1]) + dy(t[i]+h, y_a))
        y_heun.append(y_n)
    plt.plot(t, y_heun)

    plt.show()
```

```python
loeservergleich(0.1)
```

```python
loeservergleich(0.2)
```

```python
loeservergleich(0.5)
```

$$
L_i(x):=\prod_{\substack{j=0 \\ j \neq i}}^n \frac{x-x_j}{x_i-x_j}, \quad i=0, \ldots n
$$

1. Zeigen Sie, dass $\left\{L_0, \ldots, L_n\right\}$ eine Basis von $\Pi_n(\mathbb{R})$, dem Raum der Polynome auf $\mathbb{R}$ vom Grad $\leqslant n$, ist.

```python
from sympy import symbols, simplify, expand
from sympy.polys.polyfuncs import interpolate

# Define the symbol for the polynomial
x = symbols('x')

# List of points (x_i, y_i)
points = [(1, 2), (2, 3), (3, 5)]  # Example points

# Calculate the Lagrange interpolation polynomial
lagrange_poly = interpolate(points, x)

# Simplify and expand the polynomial for better readability
lagrange_poly_simplified = simplify(expand(lagrange_poly))

# Print the polynomial
print(f'Lagrange Interpolation Polynomial: {lagrange_poly_simplified}')
```

## 1.1.

Der neue Ansatz zur Lösung besteht darin sich die Begriffe der Aufgabenstellung noch einmal genau zu betrachten. 

$\Pi_n(\mathbb{R})$ ist der Raum der Polynome vom Grad $\leq \boldsymbol{n}$, die Koeffizienten aus $\mathbb{R}$ (den reellen Zahlen) haben. 

Ein Polynom in $\Pi_n(\mathbb{R})$ hat die Form: $P(x)=a_0+a_1 x+a_2 x^2+\cdots+a_n x^n,$
wobei $a_0, a_1, \ldots, a_n \in \mathbb{R}$ sind.

Der Raum $\Pi_n(\mathbb{R})$ hat die Dimension $n+1$, da ein Polynom mit Grad $\leq n$ durch $n+1$ Koeffizienten eindeutig beschrieben wird.

Eine Basis ist eine Menge von Vektoren (hier Polynome), die zwei Eigenschaften erfüllen:
1. Sie sind linear unabhängig.
2. Sie spannen den gesamten Raum auf (d. h. jedes Element des Raums kann als Linearkombination der Basisvektoren dargestellt werden).

Um hier also die Basis der Polynome auf $\mathbb{R}$ darzustellen müssen die Vektoren $a$ mit den Elementen $a_0, a_1,... a_n$ voneinander Linear Unabhängig sein. 

Die Menge der Polynome {$L_1, L_2, ... L_3$} ist das Linear unabhängig wenn die folgende Gleichung erfüllt ist:

$$c_0 L_0(x)+c_1 L_1(x)+\cdots+c_n L_n(x)=0$$

und $\forall c_i = 0$

Wenn das Lagrange-Polynom an der Stelle seiner seiner Entwicklung betrachtet wird ergibt sich das Folgende: 

$$L_i(x_i):=\prod_{\substack{j=0 \\ j \neq i}}^n \frac{x_i-x_j}{x_i-x_j} = \prod_{\substack{j=0 \\ j \neq i}}^n 1 = 1$$

Wir des aber an der Stelle eines anderen x_k betrachtet so bildet sich in einem der Faktoren der Ausdrck $x_k - x_j$ welcher natürlich gleich 0 ist in dem Fall, dass $k = j$ ist:


$$ L_i(x_k):=\prod_{\substack{j=0 \\ j \neq i}}^n \frac{x_k-x_j}{x_i-x_j}  = 0$$


c_0 L_0(x)+c_1 L_1(x)+\cdots+c_n L_n(x)=0 .

So kommen wir zu einem Ausdruck durch ein Korneckerdelta. 

$$ L_i\left(x_j\right)=\delta_{i j} = \begin{cases}1, & \text { wenn } i=j, \\
0, & \text { wenn } i \neq j .\end{cases} $$


In der Aussage vom Beginn, welche unabhängig sein sollte können wir nun $x = x_i$  mit $\in {{1,2,...,n }}$ betrachten und erhalten:
$$c_0 L_0(x)+c_1 L_1(x)+\cdots+c_n L_n(x)=0$$

$$c_0 L_0(x_0)+c_1 L_1(x_1)+\cdots+c_n L_n(x_n)=0$$

jedes $L_i$ kann nun durch 1 ersetzt werden: 

$c_0 + c_1 + \cdots + c_n = 0$

Was der Bedingung für die Unabhängigkeit entspricht.

Um nun zu zeigen, dass diese einen Basis für $\Pi_n ( \mathbb{N})$ darstellt, zeigen wir, dass sich alle Funktionen dieser Form darstellen lassen.

$$P(x)=a_0+a_1 x+a_2 x^2+\cdots+a_n x^n$$

$$P(x)=\sum_{i=0}^n c_i L_i(x)$$

Da  bereits Bewiesen wurde, dass alle $L_i$ linear unabhängig sind, braucht nun nur bewiesen werden, dass jeder Summant auch dem Grad n entspricht. 

Da dies der Ansatz des ersten Lösungsversuchs war, gilt nun: 

$$L_{n+1}(x):=\prod_{\substack{j=0 \\ j \neq n+1}}^{n+1} \frac{x-x_j}{x_{n+1}-x_j}=\prod_{j=0}^{n} \frac{x-x_j}{x_{n+1}-x_j}$$

Dieser  Ausdruck hat Faktoren von $0$ bis $n$ und somit $n+1$ Faktoren, die alle ein x enthalten, somit ist der Grad $n+1$.

$$\prod_{j=1}^{n+1} \frac{x-x_j}{x_0-x_j} = \prod_{j=1}^{n} \frac{x-x_j}{x_0-x_j} \cdot \frac{x-x_{n+1}}{x_0-x_{n+1}} $$

Bei dieser Aussage gibt es Faktoren von $1$ bis $n$ sowie einen weiteren. Also Grad n. 


Damit gilt,  dass $\left\{L_0, \ldots, L_n\right\}$ eine Basis von $\Pi_n(\mathbb{R})$, dem Raum der Polynome auf $\mathbb{R}$ vom Grad $\leqslant n$, ist.

## 2.2.

Ein System von Funktionen $\left\{f_1, f_2, \ldots, f_n\right\}$ ist orthogonal bezüglich eines inneren Produkts $\langle\cdot, \cdot\rangle$, wenn für alle Paare $i \neq j$ gilt:

$$
\left\langle f_i, f_j\right\rangle=0
$$


Das bedeutet:
- Das innere Produkt zweier verschiedener Funktionen $f_i$ und $f_j$ ist Null.
- Sie tragen keinen gemeinsamen "Anteil" zueinander bei.


Das ist die gleiche Aussage wie in der Aufgabenstellung.

$$\langle f, g\rangle_w:=\frac{2}{\pi} \int_{-1}^1 f(x) g(x) w(x) \mathrm{d} x .$$

$$ \left\langle T_m, T_n\right\rangle_w=\frac{2}{\pi} \int_{-1}^1 T_m(x) T_n(x) w(x) \mathrm{d} x .$$

$$ \left\langle T_m, T_n\right\rangle_w=\frac{2}{\pi} \int_{-1}^1 \cos (m \arccos (x)) \cos (n \arccos (x)) \cdot \frac{1}{\sqrt{1-x^2}} \mathrm{~d} x $$

An dieser Stelle kommen wir mit Integration durch Substituion weiter.

$x = \cos(\theta)$

Obergrenze:  $1 = \cos(\theta) \Rightarrow \theta = 0$

Untergrenze: $-1 = \cos(\theta) \Rightarrow \theta = \pi$

$\sin(\arccos(x)) = \sqrt{1-x^2}$

$\mathrm{~d}x = -\sin(\theta)d\theta = -\sin(\arccos(x))\mathrm{~d}\theta$

$ \left\langle T_m, T_n\right\rangle_w=\frac{2}{\pi} \int_{\pi}^0 \cos (m \arccos (x)) \cos (n \arccos (x)) \cdot \frac{1}{\sin(\arccos(x))} -\sin(\arccos(x))\mathrm{~d}\theta$

$ \left\langle T_m, T_n\right\rangle_w=\frac{2}{\pi} \int_{\pi}^0 \cos (m \theta) \cos (n \theta) \mathrm{~d}\theta$

$ \left\langle T_m, T_n\right\rangle_w=\frac{2}{\pi} \int_{0}^{\pi} \cos (m \theta) \cos (n \theta) \mathrm{~d}\theta$

Produktregel: $\cos (m \theta) \cos (n \theta)=\frac{1}{2}(\cos ((m-n) \theta)+\cos ((m+n) \theta))$

$ \left\langle T_m, T_n\right\rangle_w=\frac{2}{\pi} \int_0^\pi \cos (m \theta) \cos (n \theta) \mathrm{d} \theta= \frac{2}{\pi} \frac{1}{2} \int_0^\pi \cos ((m-n) \theta) \mathrm{d} \theta+ \frac{2}{\pi} \frac{1}{2} \int_0^\pi \cos ((m+n) \theta) \mathrm{d} \theta $

$ \left\langle T_m, T_n\right\rangle_w= \frac{1}{\pi} [\frac{1}{m-n}\sin ((m-n) \theta)]_0^\pi+ \frac{1}{\pi}  [ \frac{1}{m+n}\sin((m+n) \theta)]_0^\pi $

Falls $m = n$, so ist das innere Produkt nicht mehr Definiert. 

$ \left\langle T_m, T_n\right\rangle_w= \frac{1}{\pi} \frac{1}{m-n}\sin ((m-n)\pi)+ \frac{1}{\pi}   \frac{1}{m+n}\sin((m+n) \pi) $

Da dieser Stelle versagt der Ansatz. Von daher wird die Funktion geplottet um wenigesten zu einem Numerischen Ergebnis zu kommen. 


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def orthogonality_function(m, n):
    term1 = (1 / np.pi) * (1 / (m - n)) * np.sin((m - n) * np.pi)
    term2 = (1 / np.pi) * (1 / (m + n)) * np.sin((m + n) * np.pi)
    return term1 + term2

# Generate values for m and n
m_values = np.linspace(1, 10, 100)
n_values = np.linspace(1, 10, 100)

# Create a meshgrid for plotting
M, N = np.meshgrid(m_values, n_values)
Z = orthogonality_function(M, N)

# Plot the function
plt.figure(figsize=(8, 6))
contour = plt.contourf(M, N, Z, levels=100, cmap='viridis')
plt.colorbar(contour, label="Function Value")
plt.title("Orthogonality Function Plot")
plt.xlabel("m")
plt.ylabel("n")
plt.show()
```

# Theorieaufgabe 3 
(3 Punkte): Sei $P$ das eindeutige Polynom dritten Grades, welches die folgenden Stützstellen interpoliert
$$
\begin{array}{c|cccc}
i & 0 & 1 & 2 & 3 \\
\hline
x_i & -2 & -1 & 0 & 2 \\
\hline
f_i & -13 & 0 & 1 & 15
\end{array}
$$
Bestimmen sie $P(1)$ ohne $P$ zu berechnen und anschließend auszuwerten.


$P(1) = 2$ , wenn man die Werte von $f_i$ um eins nach unten Verschiebt wird die Punktsymmetrie sichtbar. 
Weil Freude bereitet und zur Überprüfung kommt hier nochmal der Plot:  

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data points
x_points = np.array([-2, -1, 0, 2])
f_points = np.array([-13, 0, 1, 15])

# Define the Lagrange basis polynomials
def lagrange_basis(x, i, x_points):
    result = 1
    for j in range(len(x_points)):
        if j != i:
            result *= (x - x_points[j]) / (x_points[i] - x_points[j])
    return result

# Define the interpolating polynomial P(x)
def interpolating_polynomial(x, x_points, f_points):
    P = 0
    for i in range(len(x_points)):
        P += f_points[i] * lagrange_basis(x, i, x_points)
    return P

# Generate x values for plotting
x_vals = np.linspace(-2.5, 2.5, 500)
y_vals = interpolating_polynomial(x_vals, x_points, f_points)

y_int = interpolating_polynomial(1, x_points, f_points)
# Plot the interpolating polynomial and the data points
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label="Interpolating Polynomial $P(x)$", color="blue")
plt.scatter(x_points, f_points, color="red", label="Data Points", zorder=5)
plt.axvline(1, color="green", linestyle="--", label="$x=1$")
plt.scatter(1, interpolating_polynomial(1, x_points, f_points), color="purple", label="$P(1)$", zorder=6)

# Add labels and legend
plt.title("Interpolating Polynomial and Data Points")
plt.xlabel("$x$")
plt.ylabel("$P(x)$")
plt.legend()
plt.grid()
plt.show()
print(y_int)
```
